# Visual and analytic solution comparison

```python
%matplotlib inline
```

```python
from bayesian_quadrature import BQ, bq_c
```

```python
# seed the numpy random generator, so we always get the same randomness
np.random.seed(8728)
```

```python
ntry = 10
n_candidate = 3
x_mean = 3.141592653589793
x_var = 10.0
```

```python
ns = 3
x = np.random.uniform(-3, 3, ns)
f_y = lambda x: scipy.stats.norm.pdf(x, 0, 1)
y = f_y(x)

bq = BQ(x, y, ntry, n_candidate, x_mean, x_var, s=0, h=30, w=1)
bq._fit_log_l()
bq.gp_log_l.params = (50, 5, 0)
bq._fit_l()
bq.gp_l.params = (y.max(), 1, 0)

nc = bq.x_sc.size
N = np.logspace(np.log10(30), np.log10(500), 100)
```

```python
xo = np.linspace(-10, 10, N[-1])
yo = scipy.stats.norm.pdf(xo, 0, 1)
plt.plot(xo, yo, 'k-')
plt.plot(xo, bq.l_mean(xo), 'r-')
plt.plot(bq.x_s, bq.l_s, 'ro')
```

```python
calc = np.empty(nc)
bq_c.int_K(
    calc, bq.gp_l.x[:, None],
    bq.gp_l.K.h, np.array([bq.gp_l.K.w]),
    bq.x_mean, bq.x_cov)

approx = np.empty((N.size, nc))
for i, n in enumerate(N):
    xo = np.linspace(-10, 10, n)
    approx[i] = bq_c.approx_int_K(xo, bq.gp_l, bq.x_mean, bq.x_cov)
```

```python
plt.plot(N, np.abs(calc-approx).sum(axis=-1), lw=2, color='r')
```

```python
print np.abs(calc - approx[-1]).max()
print approx[-1].sum(), calc.sum()
```

```python
plt.plot(approx[-1], lw=2)
plt.plot(calc, '--', lw=2)
```

```python
calc = bq_c.int_int_K(
    1, bq.gp_l.K.h, np.array([bq.gp_l.K.w]),
    bq.x_mean, bq.x_cov)

approx = np.empty(N.size)
for i, n in enumerate(N):
    xo = np.linspace(-10, 10, n)
    approx[i] = bq_c.approx_int_int_K(xo, bq.gp_l, bq.x_mean, bq.x_cov)
```

```python
plt.plot(N, np.abs(calc-approx), lw=2, color='r')
```

```python
print np.abs(calc - approx[-1]).max()
print approx[-1], calc
```
